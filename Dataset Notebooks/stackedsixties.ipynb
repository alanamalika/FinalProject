```python
import pandas as pd
```

```python
artworkcsv = pd.read_csv('Artworks.csv')
dates = pd.read_csv('dates.csv')
```

```python
artworkcsv['Years'] = ''
def year():
    years = []
    for i,row in artworkcsv.iterrows():
        try:
            row['DateAcquired'] == row['DateAcquired']
            annual = dates.loc[dates['Date'] == artworkcsv.loc[i,'DateAcquired'], 'Year'].values[0]
        except:
            annual = 'Unknown'
        years.append(annual)
    return years
```

```python
artworkcsv['Years'] = year()
```

```python
artworkcsv['Decades'] = ''
def decade():
    eras = []
    for i,row in artworkcsv.iterrows():
        try:
            row['Years'] == row['Years']
            tens = dates.loc[dates['Year'] == artworkcsv.loc[i,'Years'], 'Decade'].values[0]
        except:
            tens = 'Unknown'
        eras.append(tens)
    return eras
```

```python
artworkcsv['Decades'] = decade()
```
