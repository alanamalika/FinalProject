```python
import pandas as pd
```

```python
artworkcsv = pd.read_csv('Artworks.csv')
dates = pd.read_csv('dates.csv')
genfmnb = pd.read_csv('gender.csv')
```

```python
artworkcsv['Years'] = ''
def year():
    years = []
    for i,row in artworkcsv.iterrows():
        try:
            row['DateAcquired'] == row['DateAcquired']
            annual = dates.loc[dates['Date'] == artworkcsv.loc[i,'DateAcquired'], 'Year'].values[0]
        except:
            annual = 'Unknown'
        years.append(annual)
    return years
```

```python
artworkcsv['Years'] = year()
```

```python
artworkcsv['FMNB'] = ''
def gender():
    gen = []
    for i,row in artworkcsv.iterrows():
        try:
            row['Gender'] == row['Gender']
            fmnb = genfmnb.loc[genfmnb['gender2'] == artworkcsv.loc[i,'Gender'], 'gender'].values[0]
        except:
            fmnb = 'Unknown'
        gen.append(fmnb)
    return gen
```

```python
artworkcsv['FMNB'] = gender()
```
