```python
import pandas as pd
```

```python
artworkcsv = pd.read_csv('Artworks.csv')
dates = pd.read_csv('dates.csv')
nationalities = pd.read_csv('nationality-continent.csv')
genfmnb = pd.read_csv('gender.csv')
datecreate = pd.read_csv('datecreate.csv')
```

```python
artworkcsv['Year'] = ''
def year():
    years = []
    for i,row in artworkcsv.iterrows():
        try:
            row['DateAcquired'] == row['DateAcquired']
            annual = dates.loc[dates['Date'] == artworkcsv.loc[i,'DateAcquired'], 'Year'].values[0]
        except:
            annual = 'Unknown'
        years.append(annual)
    return years
```

```python
artworkcsv['Years'] = year()
```

```python
artworkcsv['Decades'] = ''
def decade():
    eras = []
    for i,row in artworkcsv.iterrows():
        try:
            row['DateAcquired'] == row['DateAcquired']
            tens = dates.loc[dates['Date'] == artworkcsv.loc[i,'DateAcquired'], 'Decade'].values[0]
        except:
            tens = 'Unknown'
        eras.append(tens)
    return eras
```

```python
artworkcsv['Decades'] = decade()
```

```python
artworkcsv['FMNB'] = ''
def gender():
    gen = []
    for i,row in artworkcsv.iterrows():
        try:
            row['Gender'] == row['Gender']
            fmnb = genfmnb.loc[genfmnb['gender2'] == artworkcsv.loc[i,'Gender'], 'gender'].values[0]
        except:
            fmnb = 'Unknown'
        gen.append(fmnb)
    return gen
```

```python
artworkcsv['FMNB'] = gender()
```
