### Top 10s:

As we discovered in the exploratory data, MoMA's collection is predominantly comprised of male artists. Only one woman made it into the top 10 lists of artists that MoMA has acquired the most artworks from. This analysis identifies the top 10 women artists that have had most artworks acquired by the institution.

```python
www = momaartwork_df.loc[momaartwork_df['Gender'] == '(Female)']
ww = www.groupby('Artist')[['Gender']].count()
w = ww.sort_values(by = 'Gender', ascending=False)
w
```

```python
t = w.head(9)
bar = t.plot(kind = 'barh')
bar.set_title('Top 10 Female Artists Most Represented at MoMA', fontsize = 16)
bar.set_xlabel('Number of Artworks Acquired',fontsize = 14)
bar.set_ylabel('Artist Name',fontsize = 14)
```
